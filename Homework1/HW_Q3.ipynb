```python
## Library Imports
import gym
import numpy as np
from stable_baselines3 import A2C
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
## Create a training environment 
env_train = gym.make('CartPole-v1')

## Create an evaluation environment
eval_env = gym.make('CartPole-v1')
```

##Basic Vanilla Advantage 

set the gae_lambda parameter to 1 <br>
set the n_step_advantage parameter to False

```python
## Initialize an Agent (Actor Critic)
vanilla_model = A2C('MlpPolicy', env_train, verbose=1, tensorboard_log ="./a2c_cartpole_tensorboard/",n_step_advantage=False)

## Training the agent for 100000 steps
vanilla_model.learn(total_timesteps=100000)
```

```python
## Evaluating the trained model
rewards_mean, rewards_std = evaluate_policy(vanilla_model, eval_env, n_eval_episodes=100)

print(f'Mean Reward: {np.round(rewards_mean,2)} +/- {np.round(rewards_std,2)}')
```

## Generalized Advantage Estimation (GAE) Implementation

set the gae lamda parameter to 0.5 <br>
set the n_step_advantage parameter to False

```python
## define the agent  
GAE_model = A2C('MlpPolicy', env_train,gae_lambda=0.5, verbose=1, tensorboard_log ="./a2c_cartpole_tensorboard/", n_step_advantage=False)

## train the agent for 100k iterations

GAE_model.learn(100000)
```

```python
## evaluate the trained agent 
rewards_mean, rewards_std = evaluate_policy(GAE_model, eval_env, n_eval_episodes=100)

print(f'Mean Reward: {np.round(rewards_mean,2)} +/- {np.round(rewards_std,2)}')
```

##Implementing the nstep advantage A2C

```python
## create the agent 
nstep_model = A2C('MlpPolicy',env_train,verbose=1,tensorboard_log="./a2c_cartpole_tensorboard/",n_step_advantage=True)

## train the agent for 100k iterations
nstep_model.learn(100000)
```

```python
## evaluate the trained agent 
rewards_mean, rewards_std = evaluate_policy(nstep_model, eval_env, n_eval_episodes=100)

print(f'Mean Reward: {np.round(rewards_mean,2)} +/- {np.round(rewards_std,2)}')
```

##Changing the hyperparameter value in GAE

```python
## define the agent  
GAE_model2 = A2C('MlpPolicy', env_train,gae_lambda=0.2, verbose=1, tensorboard_log ="./a2c_cartpole_tensorboard/", n_step_advantage=False)

## train the agent for 100k iterations

GAE_model2.learn(100000)
```

```python
## evaluate the trained agent 
rewards_mean, rewards_std = evaluate_policy(GAE_model2, eval_env, n_eval_episodes=100)

print(f'Mean Reward: {np.round(rewards_mean,2)} +/- {np.round(rewards_std,2)}')
```

For the same number of timesteps , the advantage that yielded the most rewards was the GAE advantage with gae_lambda parameter set to 0.5,When the parameter was reduced to 0.2 , the future rewards were not weighted as much and the rewards yieled for the same 100k steps was lesser

The n step advantage yielded slightly lower rewards in comparison to the vanilla and the GAE advantage. But since this advantage takes into account the other rewards it might take a longer training time to yield better results
